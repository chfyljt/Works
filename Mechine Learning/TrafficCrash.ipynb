```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("maryamrahmani/crash-analysis-system-cas-data-new-zealand")

print("Path to dataset files:", path)
```

```python
import pandas as pd

pd.set_option('display.max_rows', None)
data_path = r'{}'.format(path)
TrafficCrash_data = pd.read_csv(data_path + '/Crash_Analysis_System_(CAS)_data.csv')
TrafficCrash_data.head(1000)
```

```python
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns

TrafficCrash_data.isnull().sum()
# relation bewteen crash severity and other factors.
TrafficCrash_data_copy = TrafficCrash_data.copy()
TrafficCrash_data_copy = TrafficCrash_data_copy.apply(lambda x: x.astype('category').cat.codes if x.dtype == 'object' else x)
correlation_matrix = TrafficCrash_data_copy.corr()
significate_matrix = correlation_matrix[correlation_matrix['crashSeverity'].abs() > 0.3]
significate_matrix

# plt.figure(figsize=(10,6))
# sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
# plt.title('Correlation Matrix')
# plt.show()

# top10_correlations = correlation_matrix['crashSeverity'].abs().sort_values(ascending=False).head(11)
# top10_correlations
# le = LabelEncoder()
# for col in TrafficCrash_data_copy.select_dtypes(include='object').columns:
#     TrafficCrash_data_copy[col] = le.fit_transform(TrafficCrash_data_copy[col])

```

```python
plt.figure(figsize=(10,6))
TrafficCrash_data['speedlimitType'] = TrafficCrash_data['speedLimit'].apply(lambda x: 'Low' if x < 30 else 'Medium' if 30 <= x < 60 else 'High' if 60 <= x < 100 else 'Very High')
custom_order = ['Low', 'Medium', 'High', 'Very High']
TrafficCrash_data['speedlimitType'] = pd.Categorical(TrafficCrash_data['speedlimitType'], categories=custom_order, ordered=True)
contigency_table = pd.crosstab(TrafficCrash_data['speedlimitType'], TrafficCrash_data['crashSeverity'])
contigency_table.plot(kind='bar', stacked=True, colormap='coolwarm')
plt.title('Distribution of Crash Severity by Speed Limit')
plt.xlabel('Speed Limit Category')
plt.ylabel('Number of Crashes')
plt.xticks(rotation=0)
plt.legend(title='Crash Severity')
plt.show()
```

```python
region_order = ['Non-Injury Crash', 'Minor Crash', 'Serious Crash', 'Fatal Crash']
region_severity = TrafficCrash_data.groupby('region')['crashSeverity'].value_counts().unstack().fillna(0)
plt.figure(figsize=(10,6))
region_severity[region_order].plot(kind='bar',stacked=True, colormap='coolwarm', figsize=(12,6), width=0.8)
plt.title('Crash Severity Distribution Across Regions')
plt.xlabel('Region')
plt.ylabel('Number of Crashes')
plt.legend(title='Crash Severity')
plt.show()
```

```python
vehicle_columns = ['carStationWagon', 'vanOrUtility', 'truck']
vehicle_group = TrafficCrash_data.groupby('crashYear')[vehicle_columns].sum().reset_index()
plt.figure(figsize=(12,6))
for type in vehicle_columns:
    plt.plot(vehicle_group['crashYear'], vehicle_group[type], marker='o', label=type)
plt.xlabel("Year")
plt.ylabel("Number of Crashes")
plt.title("Trends of Vehicle Types Involved in Crashes Over Time (Smoothed)")
plt.legend()
plt.grid()
plt.show()
```
